## Q1 (1 point) — Answer below

```python
# Q1 — ANSWER CELL
FILE_PATH = "support_messages.txt"

import pandas as pd

# Load the TAB-separated file into df
df = pd.read_csv(FILE_PATH, sep="\t")

# Print df.shape and df.head(3)
print(df.shape)
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
df["word_count"] = df["message"].str.split().apply(len)
df["char_count"] = df["message"].str.len()
df["num_count"] = df["message"].str.count(r"\d")
df["upper_word_count"] = df["message"].apply(lambda x: sum(w.isupper() for w in x.split()))

print(df)
```

## Q3 (3 points) — Answer below

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)
    words = [w for w in text.split() if w not in ENGLISH_STOP_WORDS]
    return " ".join(words)

df["clean"] = df["message"].apply(clean_text)

print(df.loc[df["id"].isin([1,4]), ["message","clean"]])
```

## Q4 (2 points) — Answer below

```python
import re

df["order_id"] = df["message"].str.extract(r"(ORD-\d{4})", flags=re.IGNORECASE)
df["email"] = df["message"].str.extract(r"([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})")

print(df[["id","order_id","email"]])
```

## Q5 (1 point) — Answer below

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df["clean"])

avg = X.mean(axis=0).A1
terms = tfidf.get_feature_names_out()

top = avg.argsort()[-5:][::-1]

for i in top:
    print(terms[i], round(avg[i],4))
```
