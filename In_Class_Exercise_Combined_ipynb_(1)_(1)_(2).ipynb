```python
# Write your answer here
# Import VADER and nltk
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon (only runs once)
nltk.download('vader_lexicon')

# Initialize analyzer
sia = SentimentIntensityAnalyzer()

# Create 'compound' column
df['compound'] = df['Text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])

# Show first 5 rows of Text and compound
df[['Text', 'compound']].head()
```

### Q3. Create a sentiment label from the compound score (3 points)
Create a new column called `PredictedSentiment` using this rule:
- `compound >= 0.05` → `positive`
- `compound <= -0.05` → `negative`
- otherwise → `neutral`

Then show the first 10 rows of `Text`, `compound`, and `PredictedSentiment`.

```python
# Write your answer here
# Function to classify sentiment based on compound score
def get_sentiment_label(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Create new column
df['PredictedSentiment'] = df['compound'].apply(get_sentiment_label)

# Show first 10 rows
df[['Text', 'compound', 'PredictedSentiment']].head(10)
```

### Q4. Compare review score and sentiment (2 points)
Group by `Score` and compute the **average compound score** for each score value.

```python
# Write your answer here
avg_compound_by_score = df.groupby('Score', as_index=False)['compound'].mean()
avg_compound_by_score
```
